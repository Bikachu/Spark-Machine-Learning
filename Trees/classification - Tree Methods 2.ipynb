```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName("treeProject").getOrCreate()
```

```
data = spark.read.csv('FileStore/tables/dog_food.csv', header = True, inferSchema = True)
```

```
data.show()
```

```
from pyspark.ml.feature import VectorAssembler
```

```
data.columns
```

```
assembler = VectorAssembler(inputCols = ['A', 'B', 'C', 'D'], outputCol = 'features')
```

```
output = assembler.transform(data)
```

```
from pyspark.ml.classification import RandomForestClassifier
```

```
rfc = RandomForestClassifier(labelCol = 'Spoiled', featuresCol = 'features')
```

```
final_data = output.select('Spoiled', 'features')
```

```
rdc_model = rfc.fit(final_data)
```

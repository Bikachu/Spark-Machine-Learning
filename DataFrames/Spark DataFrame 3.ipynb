```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('aggs').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/sales_info.csv',inferSchema = True,
                    header = True)
```

```
df.show()
```

```
df.printSchema()
```

```
df.groupby("Company").min().show()


```

```
df.agg({'Sales': 'sum'}).collect()[0].asDict()
```

```
group_data = df.groupby('Company')
```

```
group_data.agg({'Sales': 'min'}).show()
```

```
from pyspark.sql.functions import *
```

```
a = df.select(stddev('Sales').alias('std'))
a.select(format_number('std',2).alias('std')).show()
```

```
df.orderBy(df['Sales'].desc()).show()
```
